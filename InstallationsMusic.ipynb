```python
from playsound import playsound
from tkinter import*

root = Tk()
root.title('Piano music sound player') #giving the title for our window
root.geometry("500x400")


def play():
	playsound('test_output4.mp3')


title=Label(root,text="Jazz Music Generator ",bd=9,relief=GROOVE,
			font=("times new roman",40,"bold"),bg="pink",fg="black")
title.pack(side=TOP,fill=X)


play_button = Button(root, text="Play music", font=("Helvetica", 32),
					relief=GROOVE, command=play)
play_button.pack(pady=20)

info=Label(root,text="Click on the button above to play music ",
		font=("times new roman",12,"bold")).pack(pady=20)
root.mainloop()
```
